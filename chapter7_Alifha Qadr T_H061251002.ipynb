```python
import numpy as np                                  # import library numpy untuk operasi numerik

def integrate_rec(f, a, b, n):                      # definisi fungsi integrasi metode rectangle (midpoint)
    h = (b - a) / n                                 # menghitung lebar tiap subinterval
    x = np.linspace(a, b, n + 1)                    # membuat array titik dari a ke b sebanyak n+1 titik
    
    area = 0                                        # inisialisasi variabel luas total
    for i in range(n):                              # perulangan sebanyak jumlah subinterval
        midpoint = x[i] + h / 2                     # menentukan titik tengah tiap subinterval
        area += f(midpoint) * h                     # menghitung luas persegi panjang dan menjumlahkannya
        
    return area                                     # mengembalikan hasil aproksimasi integral

f = lambda x: x**2                                  # mendefinisikan fungsi f(x) = x^2

a = 0                                               # batas bawah integral
b = 2                                               # batas atas integral

sup_10 = integrate_rec(f, a, b, 10)                 # menghitung integral dengan n = 10
sup_50 = integrate_rec(f, a, b, 50)                 # menghitung integral dengan n = 50
sup_100 = integrate_rec(f, a, b, 100)               # menghitung integral dengan n = 100

print('n=10  → {:.4f}'.format(sup_10))              # menampilkan hasil dengan 4 angka di belakang koma
print('n=50  → {:.4f}'.format(sup_50))              # menampilkan hasil dengan 4 angka di belakang koma
print('n=100 → {:.4f}'.format(sup_100))             # menampilkan hasil dengan 4 angka di belakang koma

exact = (b**3)/3                                    # menghitung nilai eksak integral ∫x^2 dx dari 0 ke 2
print('Nilai eksak → {:.4f}'.format(exact))         # menampilkan nilai eksak
```

```python
import numpy as np                                  # import library numpy untuk operasi numerik

def integrate_trap(f, a, b, n):                     # definisi fungsi integrasi metode trapesium
    h = (b - a) / n                                 # menghitung lebar tiap subinterval
    x = np.linspace(a, b, n + 1)                    # membuat array titik dari a ke b sebanyak n+1 titik
    
    area = h * (f(x[0]) + f(x[n])) / 2              # menghitung luas awal (ujung kiri dan kanan dengan faktor 1/2)
    for i in range(1, n):                           # perulangan untuk titik tengah (tanpa ujung)
        area += f(x[i]) * h                         # menambahkan luas tiap trapesium ke total
        
    return area                                     # mengembalikan hasil aproksimasi integral

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5)       # menghitung integral sin(x) dari 0 ke π/2 dengan n = 5
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10)     # menghitung integral sin(x) dari 0 ke π/2 dengan n = 10

print('Using n=5   → {:.4f}'.format(sup_5))         # menampilkan hasil dengan 4 angka di belakang koma
print('Using n=10  → {:.4f}'.format(sup_10))        # menampilkan hasil dengan 4 angka di belakang koma

exact = 1                                           # nilai eksak integral ∫sin(x) dx dari 0 ke π/2
print('Exact value → {:.4f}'.format(exact))         # menampilkan nilai eksak
```

```python
def simple_lithopress(z, rho=2900, g=9.8):         # fungsi untuk menghitung tekanan litostatik
    pressure = z * g * rho / 1e6                   # rumus tekanan: P = ρ g z, dibagi 1e6 untuk konversi ke MPa
    return pressure                                # mengembalikan nilai tekanan

my_pressure = simple_lithopress(z=2000)             # menghitung tekanan pada kedalaman 2000 meter

print('The pressure at 2000 meters is {0:.0f} MPa'.format(my_pressure))  # menampilkan hasil tekanan dalam MPa
```

```python
import numpy as np                                  # import library numpy untuk operasi numerik
from scipy.integrate import trapezoid               # import fungsi trapezoid untuk integrasi (tidak digunakan di kode ini)
import matplotlib.pyplot as plt                     # import matplotlib untuk visualisasi grafik

r = np.linspace(1, 6400, 6400)                      # membuat array radius dari pusat bumi (1 km sampai 6400 km)

def density():                                      # fungsi untuk membangun distribusi densitas bumi
    rho_inner_core = np.linspace(13100, 12800, 1220)   # densitas inner core (menurun)
    rho_outer_core = np.linspace(12200, 9900, 2259)    # densitas outer core
    rho_lower_mantle = np.linspace(5600, 4400, 2171)   # densitas lower mantle
    rho_upper_mantle = np.linspace(4130, 3400, 720)    # densitas upper mantle
    rho_crust = np.linspace(3100, 2700, 30)            # densitas crust
   
    rho_final = np.concatenate((                      # menggabungkan semua lapisan menjadi satu array
        rho_inner_core,
        rho_outer_core,
        rho_lower_mantle,
        rho_upper_mantle,
        rho_crust
    ))
    
    return rho_final                                 # mengembalikan distribusi densitas lengkap

rho = density()                                      # memanggil fungsi untuk mendapatkan data densitas

fig, ax = plt.subplots()                             # membuat figure dan axis untuk plot
ax.plot(r, rho, label="Densities on the Earth's Interior")  # memplot densitas terhadap radius
ax.set_ylabel(r"Density [kg/m$^3$]")                 # memberi label sumbu y
ax.set_xlabel("Distance from the Earth center r [km]")  # memberi label sumbu x
ax.legend()                                          # menampilkan legenda
ax.grid()                                            # menampilkan grid pada grafik

plt.show()                                           # menampilkan grafik ke layar
```

```python
import numpy as np                                  # import library numpy untuk operasi numerik
from scipy.integrate import trapezoid               # import fungsi integrasi metode trapesium
import matplotlib.pyplot as plt                     # import matplotlib untuk visualisasi grafik

def gravity(r):                                     # fungsi untuk menghitung percepatan gravitasi terhadap radius
    
    g = np.zeros(len(r))                            # inisialisasi array g dengan nilai nol
    Gr = 6.67408e-11                                # konstanta gravitasi universal (m^3/kg/s^2)
    r = r * 1000                                    # konversi radius dari km ke meter
    
    for i in range(1, len(r)):                      # perulangan dari indeks 1 sampai akhir
        
        r1 = r[0:i]                                 # mengambil radius dari pusat sampai titik ke-i
        rho1 = rho[0:i]                             # mengambil densitas yang bersesuaian
        r2 = r1[i-1]                                # radius pada titik terakhir interval
    
        y = rho1 * r1**2                            # membentuk fungsi integran (ρ * r^2)
        y_int = trapezoid(y, r1)                    # mengintegrasikan menggunakan metode trapesium
        
        g1 = ((4 * np.pi * Gr) / (r2**2)) * y_int   # menghitung gravitasi berdasarkan hukum gravitasi
        g[i] = g1                                   # menyimpan hasil ke array g
           
    return g                                        # mengembalikan array percepatan gravitasi

g = gravity(r)                                      # menghitung nilai gravitasi untuk setiap radius

fig, ax = plt.subplots()                            # membuat figure dan axis
ax.plot(r, g)                                       # memplot gravitasi terhadap radius
ax.grid()                                           # menampilkan grid
ax.set_ylabel(r'g [m/s$^2$]')                       # memberi label sumbu y
ax.set_xlabel('Distance from the Earth center r [km]')  # memberi label sumbu x

plt.show()                                          # menampilkan grafik
```

```python
import numpy as np                                  # import library numpy untuk operasi numerik
from scipy.integrate import trapezoid               # import fungsi integrasi metode trapesium
import matplotlib.pyplot as plt                     # import matplotlib untuk visualisasi grafik

def pressure(r, rho, g):                            # fungsi untuk menghitung tekanan sebagai fungsi radius
    
    p = np.zeros(len(r))                            # inisialisasi array tekanan dengan nol
    r = r * 1000                                    # konversi radius dari km ke meter
    
    for i in range(len(r)):                         # perulangan untuk setiap titik radius
        r1 = r[i:len(r)]                            # mengambil radius dari titik i sampai permukaan
        rho1 = rho[i:len(r)]                        # mengambil densitas pada interval tersebut
        g1 = g[i:len(r)]                            # mengambil gravitasi pada interval tersebut
        
        y = rho1 * g1                               # membentuk integran (ρ * g)
        p1 = trapezoid(y, r1)                       # integrasi tekanan menggunakan metode trapesium
        p[i] = p1                                   # menyimpan hasil tekanan pada titik i
        
    return p                                        # mengembalikan array tekanan

p = pressure(r, rho, g) / 1e9                       # menghitung tekanan dan konversi ke GPa

z = np.linspace(6400, 1, 6400)                      # membuat array kedalaman dari permukaan ke pusat bumi
fig, ax = plt.subplots()                            # membuat figure dan axis
ax.plot(z, p)                                       # memplot tekanan terhadap kedalaman
ax.grid()                                           # menampilkan grid
ax.set_ylabel('P [GPa]')                            # memberi label sumbu y
ax.set_xlabel('Depth z from the Earth Surface [km]')  # memberi label sumbu x

plt.show()                                          # menampilkan grafik
```
