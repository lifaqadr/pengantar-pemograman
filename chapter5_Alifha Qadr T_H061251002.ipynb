```python
import pandas as pd                                  # import library pandas untuk manipulasi data
from scipy.stats.mstats import gmean, hmean          # import fungsi geometric mean dan harmonic mean
import matplotlib.pyplot as plt                      # import library untuk visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file excel dan mengambil sheet tertentu

a_mean = my_dataset.Tm.mean()                        # menghitung arithmetic mean (rata-rata biasa) dari kolom Tm
g_mean = gmean(my_dataset['Tm'])                     # menghitung geometric mean dari kolom Tm
h_mean = hmean(my_dataset['Tm'])                     # menghitung harmonic mean dari kolom Tm

print ('-------')                                    # mencetak garis pemisah
print ('arithmetic mean')                            # menampilkan label arithmetic mean
print ("{0:.1f} [ppm]".format(a_mean))               # menampilkan nilai mean dengan 1 angka desimal
print ('-------')                                    # mencetak garis pemisah

print ('geometric mean')                             # menampilkan label geometric mean
print ("{0:.1f} [ppm]".format(g_mean))               # menampilkan nilai geometric mean
print ('-------')                                    # mencetak garis pemisah

print ('harmonic mean')                              # menampilkan label harmonic mean
print ("{0:.1f} [ppm]".format(h_mean))               # menampilkan nilai harmonic mean
print ('-------')                                    # mencetak garis pemisah

fig, ax = plt.subplots()                             # membuat figure dan axes untuk plotting
ax.hist(my_dataset.Tm, bins='auto', density=True, edgecolor='#FF0048', label='Measurements Hist', alpha=0.8)  # membuat histogram distribusi data Tm
ax.axvline(a_mean, color='#613613', label='Arithmetic mean', linewidth=8)   # garis vertikal untuk arithmetic mean
ax.axvline(g_mean, color='#4B382A',  label='Geometric mean', linewidth=10)  # garis vertikal untuk geometric mean
ax.axvline(h_mean, color='#997950',  label='Harmonic mean', linewidth=7)    # garis vertikal untuk harmonic mean
ax.set_xlabel('Tm [ppm]')                         # memberi label sumbu X
ax.set_ylabel('Probability density')              # memberi label sumbu Y
ax.legend()                                      # menampilkan legenda pada grafik
```

```python
import pandas as pd                                  # import library pandas untuk manipulasi data
import matplotlib.pyplot as plt                      # import matplotlib untuk visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel dan memilih sheet tertentu

variance = my_dataset['Ta'].var()                    # menghitung variance (ragam) dari kolom Ta
stddev = my_dataset['Ta'].std()                      # menghitung standard deviation (simpangan baku) dari kolom Ta

print ('-------')                                    # mencetak garis pemisah
print ('Variance')                                   # menampilkan label variance
print("{0:.0f} [square ppm]".format(variance))       # menampilkan nilai variance tanpa desimal
print ('-------')                                    # mencetak garis pemisah
print ('Standard Deviation')                         # menampilkan label standar deviasi
print("{0:.0f} [ppm]".format(stddev))                # menampilkan nilai standar deviasi tanpa desimal
print ('-------')                                    # mencetak garis pemisah

fig, ax = plt.subplots()                             # membuat objek figure dan axes untuk plotting
ax.hist(my_dataset.Ta, bins=20, density=True, edgecolor='#FF0048', label='Measurements Hist')  # membuat histogram distribusi data Ta
ax.axvline(my_dataset['Ta'].mean()-stddev, color='#D0B8A8', label=r'mean - 1$\sigma$', linewidth=8)  # garis vertikal pada mean - 1 standar deviasi
ax.axvline(my_dataset['Ta'].mean()+stddev, color='#85586F', label=r'mean + 1$\sigma$', linewidth=8)  # garis vertikal pada mean + 1 standar deviasi
ax.axvspan(my_dataset['Ta'].mean()-stddev, my_dataset['Ta'].mean()+stddev, alpha=0.1, color='#F8EDE3', label=r'mean $\pm$ 1$\sigma$')  # area arsiran antara mean ± 1 standar deviasi
ax.set_xlabel('Ta [ppm]')                            # memberi label pada sumbu X
ax.set_ylabel('Probability density')                 # memberi label pada sumbu Y
ax.legend()                                          # menampilkan legenda grafik

'''
Output:                                              # blok komentar hasil output
-------                                              # garis pemisah
Variance                                             # label variance
14021 [square ppm]                                   # nilai variance
-------                                              # garis pemisah
Standard Deviation                                   # label standar deviasi
118 [ppm]                                            # nilai standar deviasi
-------                                              # garis pemisah
'''
```

```python
import pandas as pd                                  # import library pandas untuk manipulasi data
import numpy as np                                   # import numpy untuk operasi numerik/statistik
import matplotlib.pyplot as plt                      # import matplotlib untuk visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel dan memilih sheet tertentu

q1 = np.percentile(my_dataset.Tm, 25, method='midpoint')  # menghitung kuartil pertama (Q1, persentil ke-25)
q3 = np.percentile(my_dataset.Tm, 75, method='midpoint')  # menghitung kuartil ketiga (Q3, persentil ke-75)

iqr = q3 - q1                                      # menghitung Interquartile Range (IQR = Q3 - Q1)

print ('-------')                                   # mencetak garis pemisah
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))  # menampilkan nilai IQR tanpa desimal
print ('-------')                                   # mencetak garis pemisah

fig, ax = plt.subplots()                            # membuat objek figure dan axes untuk plotting
ax.hist(my_dataset.Tm, bins='auto', density=True, edgecolor='#FF0048', label='Measurements Hist')  # membuat histogram distribusi data Tm
ax.axvline(q1, color='#B28DFF', label='Q1', linewidth=7)  # garis vertikal pada kuartil pertama (Q1)
ax.axvline(q3, color='#FFABAB', label='Q3', linewidth=7)  # garis vertikal pada kuartil ketiga (Q3)
ax.axvspan(q1, q3, alpha=0.1, color='#AFF8DB', label='Interquaritle range (IQR)')  # area arsiran antara Q1 dan Q3 (IQR)
ax.set_xlabel('Tm [ppm]')                           # memberi label pada sumbu X
ax.set_ylabel('Probability density')                # memberi label pada sumbu Y
ax.legend()                                         # menampilkan legenda grafik

'''
Output:                                             # blok komentar hasil output
-------                                             # garis pemisah
Interquaritle range (IQR): 164 [ppm]                # nilai IQR dari data
-------                                             # garis pemisah
'''
```

```python
import pandas as pd                                  # import library pandas untuk manipulasi data
import numpy as np                                   # import numpy untuk operasi numerik/statistik
import matplotlib.pyplot as plt                      # import matplotlib untuk visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel dan memilih sheet tertentu

a_mean = my_dataset.Th.mean()                        # menghitung arithmetic mean (rata-rata) dari kolom Th

median = my_dataset.Th.median()                      # menghitung median (nilai tengah) dari kolom Th

hist, bin_edges = np.histogram(my_dataset['Th'], bins=20, density=True)  # menghitung histogram dan batas bin dari data Th
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2  # menentukan modus dari bin dengan frekuensi tertinggi

fig, ax = plt.subplots()                             # membuat objek figure dan axes untuk plotting
ax.hist(my_dataset.Th, bins=20, density=True, edgecolor='#FF0048', label="Measurements Hist")  # membuat histogram distribusi data Th
ax.axvline(modal_value, color='#DE3163', label='Modal Value', linewidth=7)   # garis vertikal pada nilai modus
ax.axvline(median, color='#9F70FF', label='Median Value', linewidth=7)       # garis vertikal pada nilai median
ax.axvline(a_mean, color='#613613', label='Arithmetic mean', linewidth=7)    # garis vertikal pada nilai mean
ax.set_xlabel('Th [ppm]')                          # memberi label pada sumbu X
ax.set_ylabel('Probability density')               # memberi label pada sumbu Y
ax.legend()                                        # menampilkan legenda grafik
```

```python
import numpy as np                                  # import numpy untuk operasi numerik/statistik
from scipy.stats import skew                        # import fungsi skew untuk menghitung kemencengan distribusi

a_mean = my_dataset.Hf.mean()                       # menghitung arithmetic mean (rata-rata) dari kolom Hf
median = my_dataset.Hf.median()                     # menghitung median (nilai tengah) dari kolom Hf

hist, bin_edges = np.histogram(my_dataset['Hf'], bins=20, density=True)  # menghitung histogram dan batas bin dari data Hf
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2  # menentukan modus dari bin dengan frekuensi tertinggi

standard_deviation = my_dataset['Hf'].std()         # menghitung standar deviasi dari kolom Hf

a1 = (a_mean - modal_value) / standard_deviation    # menghitung Pearson's first coefficient of skewness
a2 = 3 * (a_mean - median) / standard_deviation     # menghitung Pearson's second coefficient of skewness
g1 = skew(my_dataset['Hf'])                         # menghitung Fisher-Pearson coefficient of skewness

print ('-------')                                   # mencetak garis pemisah
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))   # menampilkan nilai koefisien skewness pertama
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))          # menampilkan nilai koefisien skewness kedua
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))  # menampilkan nilai skewness metode Fisher
print ('-------')                                   # mencetak garis pemisah

'''
Output:                                             # blok komentar hasil output
-------                                             # garis pemisah
Pearson's first coefficient of skewness: 0.74       # nilai skewness Pearson pertama
Pearson's 2nd moment of skewness: 0.66              # nilai skewness Pearson kedua
Fisher-Pearson's coefficient of skewness: 1.26      # nilai skewness Fisher
-------                                             # garis pemisah
'''
```

```python
import pandas as pd                                  # import library pandas untuk manipulasi data
import matplotlib.pyplot as plt                      # import matplotlib untuk visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel dan memilih sheet tertentu

fig, ax = plt.subplots()                             # membuat objek figure dan axes untuk plotting

my_flierprops = dict(markerfacecolor='#161A1D', markeredgecolor='#A4161A', marker='s')  # mengatur tampilan outlier (titik ekstrem) pada boxplot
my_medianprops = dict(color='#E5383B', linewidth=3)  # mengatur tampilan garis median pada boxplot
my_boxprops = dict(facecolor='#FFD700', edgecolor='#0B090A')  # mengatur tampilan kotak utama boxplot

ax.boxplot(my_dataset.Ho, patch_artist=True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops)  # membuat boxplot untuk data Ho dengan kustomisasi
ax.set_ylabel('Ho [ppm]')                           # memberi label pada sumbu Y
ax.set_xticks([1])                                  # menentukan posisi tick pada sumbu X
ax.set_xticklabels(['all Epochs'])                  # memberi label pada sumbu X
plt.show()                                          # menampilkan grafik ke layar
```
